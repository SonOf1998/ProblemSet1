First we import OpenCV for image reading, Matplotlib for displaying and NumPy.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

We make a tuple with the file names of the 5 selected 128x128 images. The images must be in the same directory as the current ipynb file.

Then we make a list of length five for the images itself. This list will eventually contain the NumPy representation of the images (np_array with shape of (128, 128, 3) where the last dimension is the R, G and B channels respectively).

Fortunately, OpenCV's imread produces an np_array as its return value directly. We just need a little twist there. OpenCV actually reads images in BGR order, but we can easily reverse that by calling OpenCV's cvtColor (convert color) with the adequate flag.

To finish subtask 1 we just simply display the images and their tensor representation below.

```python
file_names = (
    "lomonosov128.jpg",
    "onearmhandstand128.jpg",
    "rubberduck128.jpg",
    "snitzel128.jpg",
    "switzerland128.jpg"
)


num_of_images = len(file_names)
images = [None] * num_of_images

for i in range(num_of_images):
    images[i] = cv2.cvtColor(cv2.imread(file_names[i]), cv2.COLOR_BGR2RGB)    
    plt.imshow(images[i])
    plt.axis("off")
    plt.show()
    print(images[i])
    
```

For the second subtask we make a dataset by stacking the five pictures onto the top of each other.


```python
dataset = np.vstack(images)
dataset = dataset.astype(np.float64)
```

We then calculate the mean and the standard deviation (square root of variance) for the pixels' R, G, B.

```python
mean = np.mean(dataset, axis=(0, 1))
```

```python
mean
```

```python
root_variance = np.std(dataset, axis=(0, 1))
```

```python
root_variance
```

```python
for x in range(dataset.shape[0]):
    for y in range(dataset.shape[1]):
        dataset[x, y] = (dataset[x, y] - mean) / root_variance
```

To verify that the new dataset indeed has zero as its mean and a unit standard deviation, we reevaluate variables 'mean' and 'root_variance' and expect 0 and 1 (or really close to that due to the floating point precision) as their value.

```python
mean = np.mean(dataset, axis=(0, 1))
```

For the third subtask we need the wavfile module from scipy.io to get the sample rate and to read the samples from a chosen wave file. 

With the collected data matplotlib can plot the spectrogram as requested.

P.S.: Handling an mp3 or a stereo audio file would have been much harder. **verao.wav** is a mono wave file for this reason.

```python
from scipy.io import wavfile

sample_rate, samples = wavfile.read("verao.wav")
plt.specgram(samples, Fs=sample_rate)
plt.ylabel("Frequency [Hz]")
plt.xlabel("Time [s]")
plt.show()
```

For the last exercise we need two libraries:

*urllib.request* to get the raw html content of the website and  
*bs4 / BeautifulSoup* to help us removing the html tags

```python
import urllib.request
from bs4 import BeautifulSoup
```

We read the binary content of the deep learning wiki site and then we decode it as an UTF-8 document.  
BeatifulSoup's get_text() will return the tag free version of this document.

```python
website_content_with_tags = urllib.request.urlopen("https://en.wikipedia.org/wiki/Deep_learning").read().decode("UTF-8")
bs = BeautifulSoup(website_content_with_tags)
website_content = bs.get_text()
print(website_content)
```

We initialize an empty dictionary and fill it up with letter frequencies by iterating over the website content we just extracted character by character. We limit the characters to letters only.

As most wikipedia pages has foreign characters (far-eastern, arabic, etc.) and matplotlib has hard time displaying them I suppressed warnings for this task.

```python
import warnings
warnings.filterwarnings("ignore")

letter_frequencies = {}

for char in website_content:
    if str.isalpha(char):
        char = char.lower()
        
        if char in letter_frequencies:
            letter_frequencies[char] += 1
        else:
            letter_frequencies[char] = 1


letter_frequencies = dict(sorted(letter_frequencies.items()))

plt.bar(letter_frequencies.keys(), letter_frequencies.values())
plt.show()
```
